```python
#initialize dictionary to store media pairs
media_dict = {}

for account in accounts_list:
    
    query = f"from:{account} has:images"
    
    #search all tweets from the past year
    
    start_time = datetime.fromisoformat('2018-08-10')
    tweets = client.search_all_tweets(query=query, expansions="attachments.media_keys", max_results=500, media_fields=['alt_text', 'url'], start_time=start_time)
    time.sleep(3)
    #store image url and alt_text as a tuple under the media_key
    if tweets.includes:
        for media in tweets.includes['media']:
            media_dict[media.media_key] = (media.url, media.alt_text)    
```

```python
media_dict
```

```python
import pandas as pd
SI_tweets = pd.DataFrame({'media_id': media_dict.keys(), 'img_url': [val[0] for val in media_dict.values()], 
                         'alt_text': [val[1] for val in media_dict.values()]})
```

```python
SI_tweets.dropna(inplace=True)
len(SI_tweets)
```
